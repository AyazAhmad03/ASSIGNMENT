```python
# Converts to datetime
trades["Timestamp IST"] = pd.to_datetime(
    trades["Timestamp IST"],
    format="%d-%m-%Y %H:%M"
)

sentiment["date"] = pd.to_datetime(sentiment["date"])
```

```python
trades.info()
```

```python
sentiment.info()
```

```python
#common date column for merging both dataset..
trades["date"] = trades["Timestamp IST"].dt.date
sentiment["date"] = sentiment["date"].dt.date
```

```python
merged = pd.merge(
    trades,
    sentiment,
    on="date",
    how="left"
)
```

```python
merged.shape
```

```python
merged.head()
```

```python
merged["classification"].isnull().sum()
```

```python
merged[merged["classification"].isnull()]
```

```python
merged = merged.dropna(subset=["classification"])
```

```python
merged["classification"].isnull().sum()
```

# **MARKET SENTIMENT DISTRIBUTION**

```python
import seaborn as sns

plt.figure(figsize=(8,5))

sns.countplot(data=merged, x="classification")

plt.title("Distribution of Market Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Number of Trades")
plt.xticks(rotation=45)

plt.show()
```

## Observation

* Sentiment fear has most number of trades..
* Sentiment Extreme fear has lowest number of trades..

```python
profit_sentiment = merged.groupby("classification")["Closed PnL"].mean().sort_values()

profit_sentiment
```

## AVERAGE PROFIT SENTIMENT

```python
profit_sentiment.plot(kind="bar", figsize=(8,5))

plt.ylabel("Average Closed PnL")
plt.title("Average Profit by Market Sentiment")

plt.show()
```

## Observation

* During Extreme greed sentiment traders earn most...

```python
profit=merged.groupby("classification")["Closed PnL"].sum().sort_values()
profit
```

## Total Profit by Sentiment

```python
profit.plot(kind="bar", figsize=(8,5))
plt.ylabel("Total Closed PnL")
plt.title("Total Profit by Market Sentiment")
plt.show()

```

## BUY VS SELL TRADES

## COIN DISTRIBUTION

```python
most_profitable=merged.groupby("Coin")["Closed PnL"].mean().sort_values(ascending=False).head(10)
```

```python
most_profitable.plot(kind="bar", figsize=(8,5))
plt.title("Top 10 Most Traded Coins")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    merged["Closed PnL"],
    bins=50
)

plt.title("Distribution of Closed PnL")

plt.show()
```

## CORRELATION HEATMAP

```python
corr_columns = [
    "Execution Price",
    "Size Tokens",
    "Size USD",
    "Start Position",
    "Closed PnL",
    "Fee",
    "value"          # Fear & Greed Index
]

corr = merged[corr_columns].corr()

plt.figure(figsize=(9,7))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap of Trading Features", fontsize=15, fontweight="bold")

plt.tight_layout()
plt.show()
```
